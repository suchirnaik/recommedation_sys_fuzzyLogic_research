```python
import pandas as pd
master_table = pd.read_csv('/Users/suchirnaik/Downloads/KPMG/master_table_for_fuzzy.csv')
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
master_table.head()
```

```python
df = master_table[['customer_id', 'brand']]
```

```python
customer_item_list = df.groupby('customer_id')['brand'].apply(', '.join).reset_index()
```

```python
customer_item_list
```

```python
customer_ids = master_table['customer_id'].unique()
```

```python
top_n = 5
```

```python
recommendations = {}
```

```python
# Assuming 'brand_info' is the column containing comma-separated brand names
brands_split = customer_item_list['brand'].str.split(',')
user_item_matrix = pd.get_dummies(brands_split.apply(pd.Series).stack()).sum(level=0)

# If 'brand_info' is already a list or you've split it into a list, you can directly use this:
# user_item_matrix = pd.get_dummies(master_table['brand_info'].apply(pd.Series).stack()).sum(level=0)
```

```python
user_item_matrix
```

```python
brand_similarity = cosine_similarity(user_item_matrix.T)
```

```python
brand_similarity
```

```python
similarity_df = pd.DataFrame(brand_similarity, columns=user_item_matrix.columns, index=user_item_matrix.columns)

for customer_id in customer_ids:
    # Filter data for the selected customer
    customer_data = master_table[master_table['customer_id'] == customer_id]

    # Get the brands that are most similar to the ones in the customer's purchase history
    purchased_brands = pd.get_dummies(customer_data['brand']).columns
    similar_brands = similarity_df[purchased_brands].mean().sort_values(ascending=False)

    # Exclude brands already purchased by the customer
    recommended_brands = similar_brands.index.difference(purchased_brands)[:top_n]

    recommendations[customer_id] = recommended_brands

# Print recommendations for each customer
for customer_id, recommended_brands in recommendations.items():
    print(f"Top recommendations for customer {customer_id}:")
    print(recommended_brands)
```

```python


# Load your dataset
# Replace this with the path to your dataset
data = pd.read_csv("your_dataset.csv")

# Set the list of customer IDs for which you want recommendations
customer_ids = data['customer_id'].unique()

# Define the number of top recommendations to provide
top_n = 5

# Create an empty dictionary to store recommendations for each customer
recommendations = {}

# Create a user-item matrix (brands as columns)
user_item_matrix = pd.get_dummies(data['brand'])

# Calculate brand similarity using cosine similarity for all customers
brand_similarity = cosine_similarity(user_item_matrix.T)

# Create a DataFrame to store similarity scores
similarity_df = pd.DataFrame(brand_similarity, columns=user_item_matrix.columns, index=user_item_matrix.columns)

for customer_id in customer_ids:
    # Filter data for the selected customer
    customer_data = data[data['customer_id'] == customer_id]

    # Get the brands that are most similar to the ones in the customer's purchase history
    purchased_brands = pd.get_dummies(customer_data['brand']).columns
    similar_brands = similarity_df[purchased_brands].mean().sort_values(ascending=False)

    # Exclude brands already purchased by the customer
    recommended_brands = similar_brands.index.difference(purchased_brands)[:top_n]

    recommendations[customer_id] = recommended_brands

# Print recommendations for each customer
for customer_id, recommended_brands in recommendations.items():
    print(f"Top recommendations for customer {customer_id}:")
    print(recommended_brands)
```
