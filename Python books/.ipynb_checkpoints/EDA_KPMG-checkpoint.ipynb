### A. KPMGtransactions dataset

```python
import pandas as pd
import matplotlib.pyplot as plt


def is_numerical(column):
    # Check if the column contains numerical data (integers or floating-point numbers)
    return pd.api.types.is_numeric_dtype(column)

def plot_box_plots(dataframe):
    for column_name in dataframe.columns:
        if is_numerical(dataframe[column_name]):
            plt.figure(figsize=(6, 4))
            plt.boxplot(dataframe[column_name], vert=False)
            plt.title(f'Box Plot for {column_name}')
            plt.xlabel(column_name)
            plt.show()

# Call the function to plot box plots for all numerical columns in the DataFrame
plot_box_plots(KPMGtransactions)

```

```python
postcode_counts = master_df_with_adress.groupby('postcode')['customer_id'].nunique().reset_index()
```

```python
postcode_counts
```

```python
postcode = sorted_postcode.merge(postcode_counts, on='postcode', how='inner')
```

```python
#customers per postcode
postcode.head(200)
```

```python
postcode_sorted = postcode.sort_values(by='transaction_count')
```

```python
postcode_sorted
```

```python
 # Initialize variables
previous_avg_count = None
score = 1

# Create an empty list to store the scores
scores = []

# Iterate through the sorted DataFrame
for index, row in postcode_sorted.iterrows():
    current_avg_count = row['transaction_count']
    
    # Compare with the previous average transaction count
    if current_avg_count != previous_avg_count:
        score += 1
    
    # Append the score to the scores list
    scores.append(score)
    
    # Update the previous average transaction count
    previous_avg_count = current_avg_count

# Add the 'scores' list as a new column in the sorted DataFrame
postcode_sorted['postcode_score'] = scores
```

## Rules 

##### 1. As frequency of customer increases, higher weightage will be given by the fuzzy logic.
##### 2. As the monetary value increases, higher weightage will be given.
##### 3. As the price of the product increases, higher weightage will be given.
##### 4. As the quantity of the product purchased increases, higher weightage will be given.
##### 5. Recency of 61 days will be the best. So, we should decide on a membership function which increases the membership as the customer's recency reaches close to 61 days.
##### 6. If a person does not possess a car, he should be given a higher weightage.
##### 7. Wealth segment should be sorted as High Net Worth, Mass Customer, and then Affluent Customer.
##### 8. Online order status doesn't have a great effect. We also saw from the univariate analysis that the number of online and offline orders are the same. However, if the company is trying to transition to online orders, we can give a higher weightage to offline customers.
##### 9. For Job Industry, we were not able to establish a clear demarcation.
##### 10. However, for job title, we were able to create a clear demarcation. Therefore, we will create a fuzzy logic where we make the categorical variables numerical and then give a higher membership weightage to those customers who belong to the job title which purchases more.
##### 11. We have also sorted the age by the average transaction count. Therefore, we will give weightage to the values based on the sorting.
